```python
#Importing libraries 

import csv
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd 
import requests 
import urllib.request
from time import sleep
from random import randint
import numpy as np
import nltk
from nltk import sent_tokenize
nltk.download('punkt')  
```

```python
#Empty Arrays 
news_links = []
all_claims = []
date = []
data = []
```

```python
#Accesing html 
url = 'https://thelogicalindian.com/fact-check/'

for page in range(1,90):
  urlpage = requests.get(url + str(page))
  #print(page)
  coverpage = urlpage.text
  soup = BeautifulSoup(coverpage, 'html5lib')
  #print(coverpage)

#Finding div and class 
  covers = soup.find_all('div', attrs={'class':'article_template2'})
  for i in range(0,12):
      if page >= 0:
          pagelink = 'https://thelogicalindian.com/' + covers[i].find('a')['href']
          news_links.append(pagelink)

          claim = requests.get(pagelink)
          newsclaim = claim.content
          soup_claim = BeautifulSoup(newsclaim, 'html5lib')

          body = soup_claim.find('h2', attrs={'class':'article_description'})
          
          try:
            factcheck = body.text.split(".")
            all_claims.append(factcheck[0])
            #print(factcheck[0])
          except: 
            body2 = soup_claim.find('article', attrs={'class':'tli_single_article'})
            try:
              for x in body2.find('h1'):
                factcheck2 = x.split("Fact Check:")
                all_claims.append(factcheck2[1])
                #print(factcheck2[1])
            except:
              pass
          claim_date = soup_claim.find('span', attrs={'class':'date convert-to-localtime date-only'})
          try:
            date.append(claim_date.text)
            #print(date)
          except:
            pass

          sleep(randint(2, 3))
  
print(news_links)
print(all_claims)
print(date)

          
          
```

```python
for n in range(0,len(all_claims)):
  data.append((news_links[n], all_claims[n], date[n]))

df = pd.DataFrame(data, columns=['Link','Claims','Date'])
df.head()
```

```python
df = pd.DataFrame(data, columns=['Links','Claims', 'Date'])
df.to_csv('Logical Indian Scrape.csv', index=False, encoding='utf-8')
```
